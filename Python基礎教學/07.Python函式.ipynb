## 函式的定義

### 語法：

```python
def functionName(argument1, arguement2):
    欲執行的程式區塊
    return 回傳值
````

`````python
def foo():
    return 'Hello World'
`````

`````python
foo()
`````

`````python
def foo(a, b):
    return a + b
`````

`````python
ret = foo(1, 2)
`````

`````python
ret
`````

## lambda 函式 (匿名函式)

### 語法：

```python
functionName = lambda argument1: 欲執行的程式
```

`````python
foo = lambda x, y: x+y
`````

`````python
foo(1,2)
`````

`````python
foo = lambda x: x**2
`````

`````python
foo(2)
`````

## lambda 與 [map 函數](http://www.runoob.com/python/python-func-map.html)做結合

`````python
l = range(5)
`````

`````python
list(l)
`````

`````python
# 對每個 list 內的值 +1
ret = map(lambda x: x+1,l)
`````

`````python
list(ret)
`````

`````python
# 對每個 list 內的值平方
ret = map(lambda x:x**2, [1,2,3,4])
`````

`````python
list(ret)
`````

## 函式的引數 (arguments)

* positional arguments (位置引數)
* keywords arguments (關鍵字引數)
* default arugments (預設引數)

`````python
def foo(a, b, c):
    return {'a':a, 'b':b, 'c':c}
`````

`````python
foo(1,2,3)
`````

`````python
foo(1,c=4,b=5)
`````

`````python
# 排列順序
foo(c=1,2,3)
`````

`````python
def foo(a, b=2, c=3):
    return {'a':a, 'b':b, 'c':c}
`````

`````python
foo(1,2)
`````

`````python
foo(1)
`````

`````python
# 順序
def foo(a=1, b, c=3):
    return {'a':a, 'b':b, 'c':c}
`````

`````python
def foo(*arg):
    return arg
`````

`````python
foo(1, 2, 3)
`````

`````python
foo(1, 2, 3, 4)
`````

`````python
def foo(**kwargs):
    return kwargs
`````

`````python
foo(1, 2, 3)
`````

`````python
foo(a=1, b=2)
`````

`````python
def foo(*args, **kwargs):
    print(args)
    print(kwargs)
`````

`````python
foo(1, 2, 3, a=4, b=5)
`````
